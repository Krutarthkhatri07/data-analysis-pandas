```python
# import the required libraries
import pandas as pd;
from matplotlib import pyplot as plt;
import numpy as np;
```

```python
# collect the required data from the information
countries = pd.read_csv('countries.csv');
countries_2007 = countries[countries.year == 2007];
```

```python
# plot the data through visualization
plt.scatter(countries_2007.gdpPerCapita, countries_2007.lifeExpectancy, 5);
plt.xlabel('GDP per capita');
plt.ylabel('Life Expectancy');
plt.title('GDP per capita vs. Life expectancy');
plt.show();
```

```python
a = countries_2007.gdpPerCapita.corr(countries_2007.lifeExpectancy);
```

```python
# need to plt log10(countries.gdpPerCapita) vs countries.lifeExpectancy
years_sorted = sorted(set(countries.year));

for year in years_sorted:
    data_year = countries[countries.year == year];
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy);
    plt.title(year);
    plt.ylim(0, 60000);
    plt.xlim(25, 85);
    plt.xlabel('log10 gdpPerCapita');
    plt.ylabel('Life Expectancy');
    plt.show(str(year));
    
```
