```python
# imports required libraries.
import pandas as pd;
from matplotlib import pyplot as plt;
```

```python
# arrange the necessary data.
countries = pd.read_csv('countries.csv');
usa = countries[countries.country == 'United States'];
china = countries[countries.country == 'China'];
usa_growth = usa.gdpPerCapita / usa.gdpPerCapita.iloc[0] * 100;
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100;
```

```python
plt.plot(usa.year, usa.gdpPerCapita);
plt.plot(china.year, china.gdpPerCapita);
plt.xlabel('Year');
plt.ylabel('GDP per capita');
plt.title('GDP per capita distribution of USA vs. China');
plt.legend(['USA', 'China']);
plt.show();
```

```python
plt.plot(usa.year, usa_growth);
plt.plot(china.year, china_growth);
plt.legend(['USA', 'China']);
plt.title('Growth of GDP per capita of USA vs. China');
plt.xlabel('Year');
plt.ylabel('GDP per capita');
plt.show();
```
