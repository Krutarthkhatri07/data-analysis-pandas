```python
from matplotlib import pyplot;

pyplot.plot([1, 3, 5], [1, 4, 9]); # This will plot (1, 1), (3, 4) and (5, 9) in the notebook.
pyplot.plot([3, 4, 5], [3, 6, 7]); # This will plot (3, 3), (4, 6) and (5, 7) in the notebook.

pyplot.xlabel('This is x-axis'); # This will give labels on the x and y axis.
pyplot.ylabel('This is y-axis');

pyplot.title('This is a sample plot'); # Gives title and legends for the plot.
pyplot.legend(['plot-1', 'plot-2']);

pyplot.savefig('first-plot-exported', dpi = 400); # This will export the plot as an image to 'first-plot-exported.img'
pyplot.show(); # This will show the plot in the note book.

# Note: If you have to save a figure make sure you call the savefig method before you call the show method.
# Reason: the current figure has been reset. pyplot keeps track of which figures, axes, etc are "current" 
# (i.e. have not yet been displayed with show) behind-the-scenes. So after you call 'show'.
# The figure instance is not being held onto
```
