```python
# import the required modules.

import pandas as pd;
from matplotlib import pyplot as plt;
```

```python
# consume and arrange data to compare life expectancy of all Americas and Europe in 1991.

countries = pd.read_csv('countries.csv');
countries_1997 = countries[countries.year == 1997];
americas_1997 = countries_1997[countries_1997.continent == 'Americas'];
europe_1997 = countries_1997[countries_1997.continent == 'Europe'];
```

```python
# compare the datas by plotting the histograms through subplots.

plt.subplot(2, 1, 1);
plt.title('Life expactancy distribution Americas vs. Europe');
plt.hist(americas_1997.lifeExpectancy, 15, range = (50, 80), edgecolor = 'black');
plt.ylabel('Americas');
plt.subplot(2, 1, 2);
plt.hist(europe_1997.lifeExpectancy, 15, range = (50, 80), edgecolor = 'black');
plt.ylabel('Europe');
plt.show();
```
