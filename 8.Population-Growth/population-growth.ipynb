```python
# import required libraries.
import pandas as pd;
from matplotlib import pyplot as plt;
```

```python
# collect the required data that will be used for plotting the time series.
countries = pd.read_csv('countries.csv');
usa = countries[countries.country == 'United States'];
china = countries[countries.country == 'China'];
usa_population_growth = usa.population / usa.population.iloc[0];
china_population_growth = china.population / china.population.iloc[0];
```

```python
# plot the non time series comparision via a line chart.
plt.plot(usa.year, usa.population);
plt.plot(china.year, china.population);
plt.title('Population of USA vs. China');
plt.xlabel('Year');
plt.ylabel('Population');
plt.legend(['USA', 'China']);
plt.show();
```

```python
# plot the time series (relative) comparision via a line chart. This helps more in knowing the growth rate of y in x units.
plt.plot(usa.year, usa_population_growth);
plt.plot(china.year, china_population_growth);
plt.xlabel('Year');
plt.ylabel('Population Growth');
plt.title('Population growth rate of USA vs. China.');
plt.legend(['USA', 'China']);
plt.show();
```
